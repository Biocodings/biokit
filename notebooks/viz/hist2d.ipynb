# wrapping of pylab hist2d function.

```python
from biokit import viz
%pylab inline
```

```python
X = pylab.randn(10000)
Y = pylab.randn(10000)
```

```python
import pandas as pd
df = pd.DataFrame({'X':X, 'Y':Y})
```

```python
from biokit.viz import hist2d
h = hist2d.Hist2D(df)
res = h.plot(bins=[40,40], contour=True, nnorm='log', 
             Nlevels=6)
```
